```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
titanic.describe()
```

```python
titanic["Age"].mean()
```

```python
titanic.dropna(subset=["Embarked"])
titanic["Cabin"].fillna("Unknown")
titanic["Age"].fillna(titanic["Age"].mean())
```

```python
titanic.isnull().sum()
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(titanic["Age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=titanic, x="Sex", hue="Survived", palette="mako")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", loc="upper right")

points = [
    "This plot shows the survival counts by gender on the Titanic.",
    "The 'Survived' legend represents whether a passenger survived (1) or not (0).",
    "Graph clearly shows that female passengers are survived more than male passengers."
]

for i, point in enumerate(points):
    plt.text(0, -120 - 100*i, f"{i+1}. {point}", ha='left', va = 'top' , fontsize=10, wrap=True)

plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data=titanic, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")

points = [
    "This plot shows the survival counts by ticket fare on the Titanic.",
    "The 'Survived' legend represents whether a passenger survived (1) or not (0).",
    "Graph clearly shows that passengers whose tickets are expensive are survived more than less expensive tickets."
]

for i, point in enumerate(points):
    plt.text(0, -150 - 100*i, f"{i+1}. {point}", ha='left', va = 'top' , fontsize=10, wrap=True)

plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.barplot(data=titanic, x="Pclass", y="Age", hue="Survived")
plt.title("Bar Plot of Passenger Class vs Age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Survived")

points = [
    "This plot shows the survival counts by age and passenger class on the Titanic.",
    "The 'Survived' legend represents whether a passenger survived (1) or not (0).",
    "Graph clearly shows that passengers of class 1 are survived more than class 2 & 3.",
    "Also passengers who are older are able to survive easily compare to younger one."
]

for i, point in enumerate(points):
    plt.text(0, -10 - 10*i, f"{i+1}. {point}", ha='left', va = 'top' , fontsize=10, wrap=True)

plt.show()
```
